```python
# Hypothesis Statement
# h0(Null hypothesis) = All proportions are equal.
# hA(Alternate hypothesis) = Not all proportions are equal. 
# Assuming Î± = 0.05
```

```python
#IMPORTING ALL THE REQUIRED LIBRARIES
import pandas as pd
import numpy as np
import scipy
from scipy import stats
from scipy.stats import chi2_contingency
```

```python
# loading the dataset
buyer = pd.read_csv('../input/buyer-ratio/BuyerRatio.csv')
```

```python
# viewing the data set
buyer
```

```python
# will display the information about the data set i.e. datatypes, no null/null values, index range
buyer.info()
```

```python
# making array
d_array = np.array([[50,142,131,70],[435,1523,1356,750]])
```

```python
# performing the test(relevant to the data set provided - chi2 test)
# (since there is discrete data present we are going ahead with chi2 test)
p_value = chi2_contingency(d_array)
```

```python
# Viewing the p-value
p_value
```

```python
# The p-value = 0.66 > 0.05 therefore the null hypothesis is correct, 
# We will accept the null hypothesis.
# i.e. All proportions are equal. 
```
