```python
# making array
d_form_array = np.array([[271,267,269,280],[29,33,31,20]])
```

```python
# performing the test(relevant to the data set provided - chi2 test)
# (since there is discrete data present we are going ahead with chi2 test)
p_value = chi2_contingency(d_form_array)
```

```python
# Viewing the p-value
p_value
```

```python
# The p-value = 0.2771 which is subsequently greater than the value of α=0.05
# Therefore the null hypothesis is true
# i.e. the defect varies by centre
```
