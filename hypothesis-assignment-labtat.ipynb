```python
# Hypothesis Statement
# h0(Null hypothesis) = There is no difference b/n the average (TAT) reports of the laboratories.
# hA(Alternate hypothesis) = There is difference in the average (TAT) reports of the laboratories. 
# Assuming Î± = 0.05
```

```python
#IMPORTING ALL THE REQUIRED LIBRARIES
import pandas as pd
import scipy 
from scipy import stats
import numpy as np
```

```python
# loading the dataset
labtat = pd.read_csv('../input/labtat/LabTAT.csv')
```

```python
# performing the test(relevant to the data set provided - Annova-test) 
stats.f_oneway(labtat.iloc[:,0], labtat.iloc[:,1], labtat.iloc[:,2], labtat.iloc[:,3]) 
```

```python
# p-value = 2.1156708949992414e-57 which is significantly less than 0.05, therefore accepting the alternate hypothesis to be true.
# i.e. There is difference in the average (TAT) reports of the laboratories.
```
